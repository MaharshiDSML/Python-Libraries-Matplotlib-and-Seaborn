## <code style="background:yellow;color:black">1. Line Chart:</code>

* **<code style="background:yellow;color:black">A line chart,</code> also known as a line graph or line plot, is a type of data visualization that displays data points as individual markers connected by straight lines. In other words, it shows the relationship between two variables, x and y.** 
* **It is particularly useful for showing trends and changes in data over a continuous interval or time series.**
* Line charts are commonly used in various fields, including business, science, and finance, to illustrate how data values evolve over time or in a continuous sequence. 
* Line charts are especially effective for showing trends, patterns, and fluctuations in data. They make it easy to identify relationships, see how data values increase or decrease over time, and identify outliers or anomalies.

**<code style="background:yellow;color:black">Key characteristics of a line chart include:**</code>

* **X-Axis:** The horizontal axis represents the independent variable, typically time or a continuous numeric variable.

* **Y-Axis:** The vertical axis represents the dependent variable, which could be any numeric value or quantity you want to measure or track.

* **Data Points:** Each data point is represented as a marker (usually a dot or a small symbol) at a specific x-coordinate and y-coordinate on the chart.

* **Lines:** The data points are connected by straight lines, creating a continuous visual representation of how the data changes over the interval of the x-axis.

**<code style="background:yellow;color:black">Common use cases for line charts include:**</code>

* **Stock Price Trends:** Line charts are frequently used to track the historical prices of stocks or financial instruments.

* **Sales and Revenue Trends:** Businesses use line charts to visualize sales or revenue over different time periods, helping them identify seasonal trends and make informed decisions.

* **Temperature and Weather Data:** Meteorologists use line charts to display temperature and weather conditions over time, allowing them to spot long-term trends and seasonal variations.

* **Scientific Data:** Researchers use line charts to display data from experiments, observations, and measurements to understand how variables change under specific conditions.

* **Website Traffic Analytics:** Line charts can be used to visualize website traffic metrics over time, such as daily or monthly page views, bounce rates, or user engagement.

Creating a line chart typically involves using software tools like Excel, Python libraries (e.g., Matplotlib or Seaborn), or dedicated data visualization tools. You provide the data points, and the software or tool connects them with lines to create the chart. Line charts are straightforward to read and are a valuable tool for conveying temporal trends and patterns in data.

## <code style="background:yellow;color:black">2. Bar Chart:</code>

* **<code style="background:yellow;color:black">A bar chart,</code> also known as a bar graph, is a common type of data visualization that represents data using rectangular bars or columns.**
* **Bar charts are particularly effective for displaying and comparing discrete categories or groups and the quantities associated with each category.**
* They provide a visual means of understanding the relationships and differences between data points in a straightforward and easily interpretable manner.
* Bar charts can be presented in various orientations, including horizontal and vertical. 
* In a vertical bar chart, the bars run from the bottom to the top of the chart, while in a horizontal bar chart, they run from left to right. The choice of orientation depends on the data and the presentation style.

**<code style="background:yellow;color:black">Key characteristics of a bar chart include:**</code>

* **X-Axis:** The horizontal axis typically represents categories or groups, such as different products, months, countries, or any distinct labels.

* **Y-Axis:** The vertical axis represents the values or quantities associated with each category. This can be any numeric value, such as counts, amounts, percentages, or other measures.

* **Bars:** Each category is represented by a rectangular bar that extends from the X-axis to a height determined by the corresponding value on the Y-axis.

**<code style="background:yellow;color:black">Common use cases for bar charts include:**</code>

* **Comparison of Categories:** Bar charts are excellent for comparing different categories by displaying their values side by side. For example, you can compare sales figures for different products, performance ratings for different employees, or revenue from different regions.

* **Ranking and Sorting:** Bar charts can help visualize the ranking or sorting of items based on their values. The length of the bars provides a clear indication of which category has the highest or lowest value.

* **Categorical Data:** Bar charts are suitable for visualizing categorical data, such as survey responses, types of products, or demographic groups.

* **Frequency Distributions:** Bar charts can be used to display the frequency of data within specific categories or bins, creating histograms.

Creating a bar chart can be done using a variety of software tools, including spreadsheet applications like Excel, data visualization libraries in programming languages like Python (e.g., Matplotlib or Seaborn), or dedicated data visualization software. You provide the data for the categories and their associated values, and the software generates the bar chart, making it a valuable tool for conveying comparisons and insights from categorical data.

## <code style="background:yellow;color:black">3. Scatter Plot:</code>

* **<code style="background:yellow;color:black">A scatter plot,</code> also known as a scatter chart or scattergram, is a type of data visualization that displays individual data points on a two-dimensional graph.**
* **It shows us a bunch of data scattered in an area.**
* It is used to represent the relationships or correlations between two variables, helping to identify patterns, trends, clusters, or outliers in the data. 
* Each data point is plotted as a single point on the graph, with one variable represented on the X-axis and the other on the Y-axis.

**<code style="background:yellow;color:black">Key characteristics of a scatter plot include:**</code>

* **X-Axis:** The horizontal axis represents one variable, often referred to as the independent variable. This variable is typically the one you control or that you believe influences the other variable.

* **Y-Axis:** The vertical axis represents the other variable, often referred to as the dependent variable. This variable is the one you are trying to understand or predict.

* **Data Points:** Each data point is represented as a single point on the graph, with its position determined by the values of the independent and dependent variables. Data points can be marked with dots, symbols, or shapes.

**<code style="background:yellow;color:black">Scatter plots are particularly useful for the following purposes:**</code>

* **Correlation Analysis:** Scatter plots are a powerful tool for determining the strength and nature of the relationship between two variables. Positive correlations, negative correlations, or no correlations can be identified visually.

* **Outlier Detection:** Outliers, which are data points that significantly deviate from the general trend, are easily spotted in a scatter plot.

* **Pattern Recognition:** Scatter plots can reveal patterns in data, such as clusters, trends, or nonlinear relationships, which might not be evident in other types of graphs.

* **Comparison of Data Points:** Scatter plots allow for the direct comparison of data points, making it easy to see how one variable's value corresponds to the other.

* **Prediction:** When used in combination with regression analysis, scatter plots can help predict one variable's value based on the other variable.

Scatter plots are widely used in fields like statistics, data analysis, scientific research, and social sciences to explore and communicate relationships between variables. They offer a visual way to understand and interpret the data, which can be especially valuable when studying patterns and associations.

## <code style="background:yellow;color:black">4. Histogram:</code>

* **<code style="background:yellow;color:black">A histogram</code> is a graphical representation of the distribution of a dataset.** 
* **It displays the frequency or count of data points falling within specific intervals or "bins."**
* Histograms are particularly useful for understanding the underlying patterns and characteristics of a dataset, including its central tendency, variability, and shape.

**<code style="background:yellow;color:black">Key characteristics of a histogram include:**</code>

* **X-Axis:** The horizontal axis, also known as the "bin" or "interval" axis, represents the range of values in the dataset divided into bins or intervals. Each bin is a range of values that data points can fall into.

* **Y-Axis:** The vertical axis represents the frequency or count of data points within each bin. It indicates how many data points fall within each interval.

* **Bars:** The histogram is constructed by drawing rectangular bars that correspond to each bin's frequency. The height of each bar represents the count of data points in that bin, and the width of the bars is typically constant between bins.

**<code style="background:yellow;color:black">Histograms are particularly useful for the following purposes:</code>**

* **Data Distribution:** Histograms provide a visual representation of the distribution of data, showing whether the data is symmetric, skewed, bimodal, or has other characteristics.

* **Central Tendency:** You can identify the central tendency of the data by looking at the peak of the histogram. In a normal distribution, for instance, the peak corresponds to the mean or median.

* **Variability:** The spread or variability of data is reflected in the width of the distribution. A wider distribution indicates greater variability, while a narrower one suggests less variability.

* **Outlier Detection:** Outliers or extreme values can often be identified as data points that fall outside the typical range of values represented by the bulk of the histogram.

* **Data Segmentation:** Histograms can be used to segment data into different categories or ranges, making it easier to see where most of the data falls.

**<code style="background:yellow;color:black">Creating a histogram typically involves the following steps:</code>**

* Define the range of values to be covered by the histogram, known as the bin edges or intervals.
* Divide the data into the specified intervals and count the number of data points that fall into each interval.
* Create rectangular bars on the histogram, with the height of each bar corresponding to the frequency count in the respective interval.

Histograms are widely used in statistics, data analysis, and various scientific disciplines to explore the distribution of data, identify patterns, and make data-driven decisions. They are a fundamental tool for understanding the characteristics of datasets, especially in cases where the distribution of data is of interest.

## <code style="background:yellow;color:black">5. Pie Chart:</code>

* **<code style="background:yellow;color:black">A pie chart</code> is a circular data visualization that is used to represent data in a way that shows the proportions of different categories or components within a whole.**
* It is called a "pie chart" because it looks like a sliced pie with each slice representing a portion of the whole. Each slice's size or angle is proportional to the quantity it represents in relation to the entire dataset.
* Pie charts are best suited for visualizing data where we want to show how various categories or components contribute to a whole and understand their relative proportions. They are useful for providing a quick and intuitive understanding of data distribution.

**<code style="background:yellow;color:black">Key characteristics of a pie chart include:**</code>

* **Circle:** A pie chart is typically a circular chart, where the entire circle represents 100% or the whole dataset.

* **Slices:** The circle is divided into slices or sectors, with each slice representing a different category or component of the data. The size of each slice is determined by the proportion or percentage of the whole that it represents.

* **Labels:** Labels are often placed outside each slice, indicating the category or component it represents and, sometimes, the exact percentage or value it corresponds to.

**<code style="background:yellow;color:black">Common use cases for pie charts include:**</code>

* **Market Share:** Pie charts can represent the market share of different companies or products within an industry.

* **Budget Allocation:** They can show how a budget is allocated to different expense categories.

* **Survey Responses:** Pie charts are often used to display the distribution of survey responses in various categories or answer choices.

* **Composition of a Whole:** They are useful for displaying the composition of a population by age groups, demographics, or any other categorical breakdown.

* **Proportions or Percentages:** When you want to emphasize percentages or proportions, pie charts provide a visual representation of these values.

It's important to note that pie charts have limitations, particularly when there are many categories or when the differences between categories are subtle. In such cases, a bar chart or other chart types might be more suitable for displaying the data clearly.

## <code style="background:yellow;color:black">6. Density Plot:</code>

* **<code style="background:yellow;color:black">A density plot,</code> also known as a kernel density plot or kernel density estimate (KDE), is a data visualization technique used to estimate and display the probability density function of a continuous random variable.**
* It provides a smooth representation of the underlying probability distribution of a dataset, showing where data is most concentrated and how it is spread out.

**<code style="background:yellow;color:black">Key characteristics of a density plot include:**</code>

* **Smooth Curves:** Density plots use smooth curves or lines to represent the estimated probability density. These curves are typically unimodal (one peak) and can reveal the shape and characteristics of the data distribution.

* **X-Axis:** The horizontal axis represents the range of values of the variable being analyzed, and it's continuous.

* **Y-Axis:** The vertical axis represents the estimated probability density, showing the relative likelihood of values within each interval.

* **Bandwidth Parameter:** The bandwidth parameter controls the smoothness of the KDE plot. A smaller bandwidth results in a more detailed, wiggly plot, while a larger bandwidth leads to a smoother, broader plot.

* **Normalized Density:** The area under the KDE curve integrates to 1, making it a probability density function. This allows you to understand the likelihood of data points falling in different regions of the data space.

**<code style="background:yellow;color:black">Density plots are valuable for various purposes, including:**</code>

* **Visualizing Data Distribution:** They provide a smooth representation of the data distribution, making it easier to observe patterns, modes, and deviations.

* **Comparison:** You can create density plots for different groups or categories and compare their distributions to understand differences or similarities.

* **Outlier Detection:** Unusual or extreme data points can be spotted in density plots where they deviate from the primary distribution.

* **Data Analysis:** Density plots are often used in statistical analysis, particularly when you want to assess the shape and characteristics of data distributions.

Density plots are especially useful when dealing with datasets containing a large number of data points or continuous data, such as measurements, scores, or values that don't fit neatly into discrete categories. They provide a way to visualize the underlying probability distribution, revealing insights about the data's central tendency, variability, and potential skewness.

## <code style="background:yellow;color:black">7. Box Plot:</code>

* **<code style="background:yellow;color:black">A box and whisker plot, also known as a box plot,</code> is a data visualization that provides a graphical summary of the distribution of a dataset.**
* It is particularly useful for displaying the spread and central tendency of the data, as well as identifying potential outliers. 
* Box and whisker plots are commonly used in statistics and data analysis.
* Box and whisker plots are especially effective when we want to visualize the distribution of data and understand the variability within the dataset, especially when dealing with multiple groups or categories.

**<code style="background:yellow;color:black">A typical box and whisker plot consists of the following components:</code>**

* **Box**: The box represents the interquartile range (IQR), which includes the middle 50% of the data. It is drawn as a rectangular box that spans from the first quartile (Q1) to the third quartile (Q3). The length of the box corresponds to the IQR.

* **Median:** A horizontal line or bar inside the box indicates the median, which is the value that divides the data into two equal halves when ordered.

* **Whiskers:** The whiskers extend from the edges of the box to the minimum and maximum data values within a defined range. The whiskers show the data's overall spread.

* **Outliers:** Individual data points located outside the whiskers are considered outliers and are often displayed as individual data points. Outliers are values that significantly deviate from the main distribution of the data.

**<code style="background:yellow;color:black">Box and whisker plots are valuable for various purposes:</code>**

* **Distribution Visualization:** They provide a visual representation of the spread and central tendency of a dataset, helping you to understand its shape and characteristics.

* **Outlier Detection:** Outliers are easily identified as data points that fall outside the whiskers, making them useful for spotting extreme values.

* **Data Summary:** Box and whisker plots offer a compact summary of the data distribution, including the range, quartiles, and median, making it easy to compare different datasets.

Creating a box and whisker plot involves calculating the quartiles and median of the dataset and then constructing the plot using these values. They are often created using data visualization tools, software, or programming libraries, which can automatically generate the plot based on your data.

## <code style="background:yellow;color:black">EDA and ExDA:</code>

**Exploratory Data Analysis (EDA) and Explanatory Data Analysis (ExDA) are two phases of the data analysis process, each serving distinct purposes in understanding and communicating insights from a dataset. In other words,** 

**<code style="background:yellow;color:black">Exploratory -> Observe patterns by crunching numbers!**</code><br>
**<code style="background:yellow;color:black">Explanatory -> Presentation of insights using visuals!**</code>

* Now, everything that we have learned so far, crunching numbers, creating tables, creating groupby queries and finding insights all of that falls under EDA. 
* And after crunching data, we have to present our data and its insights using visuals. So, from now on we will see, explore and learn ExDA.

**<code style="background:yellow;color:black">Exploratory Data Analysis (EDA):**</code>

* **Purpose:** EDA is the initial phase of data analysis, primarily focused on summarizing and visualizing the data to gain a deeper understanding of its characteristics. It's about exploring the dataset to discover patterns, anomalies, relationships, and potential insights.

* **Methods:** In EDA, data analysts or scientists use various techniques to uncover the structure of the data, including summary statistics, data visualization (such as histograms, scatter plots, box plots, and density plots), and exploration of relationships through correlation analysis. EDA often involves data cleaning and preprocessing to handle missing values or outliers.

* **Audience:** The primary audience for EDA is typically the data analyst or scientist themselves. EDA helps them formulate hypotheses, identify variables of interest, and prepare for more in-depth analyses.

* **Tools:**  Common tools used in EDA include data analysis libraries like Python's Pandas and visualization libraries like Matplotlib and Seaborn.

**<code style="background:yellow;color:black">Explanatory Data Analysis (ExDA):**</code>

* **Purpose:** ExDA is the phase of data analysis that comes after EDA and focuses on presenting and communicating the findings and insights derived from the data to a broader audience. It's about telling a story and explaining the significance of the analysis.

* **Methods:** ExDA typically involves creating clear and informative visualizations, creating reports or presentations, and writing narratives that communicate the results in a coherent and understandable way. The emphasis is on making the findings accessible to a non-technical audience.

* **Audience:** The primary audience for ExDA includes stakeholders, decision-makers, clients, or anyone who needs to understand the implications of the data analysis. The goal is to convey the insights in a way that is persuasive and actionable.

* **Tools:** ExDA often employs data visualization libraries and tools like Matplotlib, Seaborn, or specialized tools for generating interactive visualizations and dashboards (e.g., Tableau, Power BI).

In summary, EDA is about exploring and understanding the data, whereas ExDA is about presenting and explaining the insights derived from the data analysis. Both phases are integral to the data analysis process, as they help in not only understanding the data but also in effectively communicating and utilizing the findings to make informed decisions or draw conclusions.

## <code style="background:yellow;color:black">Data:</code>

**<code style="background:yellow;color:black">Data comes in various forms and types,</code> and these differences are important to consider when working with data for analysis and decision-making. Here are some common types of data:**

* **<code style="background:yellow;color:black">Categorical Data or Discrete Data:</code> Categorical data represents discrete categories or labels. It can be divided into two subtypes:**

    * **<u>Non-Ordinal or Nominal Data:</u>** Categories without a natural order or ranking, like color of fruits or gender like male and female.
    * **<u>Ordinal Data:</u>** Categories with a specific order or ranking, such as education levels (e.g., high school, bachelor's, master's) or levels like low, medium and high. <br><br>

* **<code style="background:yellow;color:black">Numerical Data or Continuous Data:</code> Numerical data represents values that can be measured and can take on numeric values. It can be further divided into:**

    * **<u>Continuous Data:</u>** Data that can take on any value within a range. Examples include height, weight, and temperature.
    * **<u>Discrete Data:</u>** Data that can only take specific, distinct values. Examples include the number of people in a household or the count of customer visits.<br><br>

* **<code style="background:yellow;color:black">Time Series Data:</code>** Time series data consists of observations collected over a sequence of time intervals. It's commonly used to analyze data that evolves over time, such as stock prices, weather data, or daily sales figures.

* **<code style="background:yellow;color:black">Text Data:</code>** Text data includes textual information, such as documents, articles, tweets, or reviews. Analyzing text data often involves natural language processing (NLP) techniques.

* **<code style="background:yellow;color:black">Geospatial Data:</code>** Geospatial data relates to geographic locations and can include latitude, longitude, addresses, or information about geographical features. It's used in applications like mapping, GPS navigation, and spatial analysis.

These are some of the key types of data, and data can be further categorized and transformed based on its characteristics and usage in different domains. Understanding the type of data you're working with is essential for choosing the appropriate analytical methods and tools.

**Figure:** Figure refers to the  entire area or canvas where one or more plots or subplots are created. The **overall window** or page that everything is drawn on, sort of like a dashboard we have in tableau.  It is the top-level container that holds all elements of a data visualization, such as axes, titles, labels, and legends. A figure can contain one or multiple plots (subplots) arranged in rows and columns. The figure can have a super title called "Suptitle". 
  - You can create multiple independent Figures in Jupyter.
  - If you run the code in terminal, separate windows will pop-up

**Axes:** To the figure we can add multiple **Axes** which represents a plot. In other words, **Axes** refers to the individual **plotting area or region** within a figure where we create and display our data visualizations or on which the **data is plotted** with functions such as `plot()`. An axes object represents a single plot with its own coordinate system, allowing us to control aspects like scales, labels, titles, and data plotting within that specific area.

**Axis**: Simply the `x-axis` and `y-axis`

**x-label**: Name of x-axis

**y-label**: Name of y-axis

**Major ticks**: subdivides the axis into major units. They appear by default during plotting.

**Minor ticks**: subdivides the major tick units. They are by default hidden and can be toggled on.

**Title**: Title of each plot **(Axes)**, giving information about the same.

**Legend**: describes the elements in the plot, blue and green curves in this case

**Suptitle**: The common title of all the plots


These are the major components of a matplotlib plot!

### <code style="background:yellow;color:black">bar() function:</code>

**The <code style="background:yellow;color:black">bar() function</code> in matplotlib is used to create bar charts or bar plots.** The bar() function is a versatile tool that can be used for creating both vertical and horizontal bar charts, stacked bar charts, and grouped bar charts. We can further customize the appearance of the bars and the plot to suit your specific data visualization needs.

**<code style="background:yellow;color:black">Here are some of the common arguments you can pass to the bar() function:**</code>

* **x:** The x-coordinates of the left sides of the bars. This argument is usually a sequence of values representing the positions of the bars on the x-axis.

* **height:** The heights of the bars, which determine their vertical size. This argument is typically a sequence of values corresponding to the heights of each bar.

* **width:** The width of the bars. This argument controls the width of the bars and is usually a scalar value.

* **bottom:** The y-coordinate of the bottom of the bars. It's an optional argument that can be used to create stacked bar charts.

* **align:** This argument specifies the alignment of the bars with respect to the x-coordinates. Common values include 'center' (default), 'edge', and 'centered'.

* **color:** The color of the bars. You can specify a single color or a list of colors to customize the colors of individual bars.

* **edgecolor:** The color of the bar edges. You can set this to define the color of the border lines around the bars.

* **linewidth:** The width of the border lines around the bars. You can use this to adjust the thickness of the bar edges.
<br>
* **In the example below, index of genre_counts represents the labels for the bars, and values represent the heights or values of the bars.**
* We pass these data as arguments to the bar() function, and it will generate a bar chart displaying these values. 
* We can customize various aspects of the bar chart, such as the colors, width, and more using additional arguments and settings.

### <code style="background:yellow;color:black">figure() function:</code>

**The <code style="background:yellow;color:black">figure() function</code> is used to create a new figure or clear an existing figure. The figure() function in Matplotlib can take several optional arguments to customize the properties of the figure. Here are some of the common arguments you can pass to figure():**

* **num:** This argument specifies the figure number. If we don't specify a number, Matplotlib will automatically assign one. If we want to update or refer to a specific figure later in our code, you can provide a number as an argument.

* **figsize:** This argument allows us to set the width and height of the figure in inches. It is a tuple of two floats, e.g., figsize=(6, 4).

* **dpi:** This argument sets the dots per inch (resolution) of the figure. The default value is usually 100. We can increase it to improve the image's quality.

* **facecolor:** We can specify the background color of the figure using this argument. The color can be a string (e.g., 'white') or an RGB tuple (e.g., (0.8, 0.8, 0.8)).

* **edgecolor:** This argument defines the color of the border around the figure. It accepts color names or RGB tuples.

* **frameon:** A boolean value that controls whether the figure frame (border) is visible.

* **clear:** If set to True, this argument will clear the figure when called. If set to False, it will not clear the existing figure.

**In the example, below we are using figsize argument to create a bigger figure and it automatically solves our problem.**

```python
plt.figure(figsize = (12, 8))

plt.bar(x_bar, y_bar)
```

### <code style="background:yellow;color:black">xticks() function:</code> 

**The <code style="background:yellow;color:black">xticks() function</code> in Matplotlib is used to customize the tick locations and labels on the x-axis of a plot. It allows us to set specific positions for ticks and define custom labels for those tick positions.**

### <code style="background:yellow;color:black">yticks() function:</code>

**The <code style="background:yellow;color:black">yticks() function</code> in Matplotlib is used to customize the tick locations and labels on the y-axis of a plot. Similar to xticks(), it allows us to set specific positions for ticks and define custom labels for those tick positions on the y-axis.**

* **The <code style="background:yellow;color:black">xticks() function and yticks() function</code> in Matplotlib takes several optional arguments that allow you to customize the appearance of the x-axis ticks.** 
* Here are the common arguments we can pass to xticks() or yticks():

    * **ticks:** This argument is a list or array of tick locations where we want the ticks to appear on the x-axis in case of xticks() or y-axis in case of yticks().

    * **labels:** This argument is an optional list or array of tick labels. If provided, it should have the same length as the ticks argument, and it will label the ticks accordingly.

    * **fontdict:** This argument is a dictionary specifying font properties for the tick labels. We can set attributes like 'family', 'size', 'weight', 'style', and others to control the appearance of the tick labels.

    * **rotation:** This argument allows us to specify the rotation angle (in degrees) for the tick labels on the x-axis in case of xticks() or y-axis in case of yticks(). It is used to control the orientation of the tick label text. Positive values rotate the text counterclockwise, while negative values rotate it clockwise.

    * **fontsize:** This argument lets us set the font size for the tick labels on the x-axis in case of xticks() or y-axis in case of yticks(). We can specify the font size as an integer or a float, which represents the size in points.

**Here in the code below, we are passing rotation nad fontsize arguments in xticks() function.**

### <code style="background:yellow;color:black">show() function:</code>

**The <code style="background:yellow;color:black">show() function</code> is used to display the currently created plot or figure on the screen. It essentially renders the plot that we've created and makes it visible in a graphical user interface (GUI) window or, if we're using Matplotlib in a Jupyter Notebook or IPython environment, within the notebook itself.**

* After calling show(), a window or output within the Jupyter Notebook will appear, showing the plot. 
* We can interact with the plot, zoom in/out, save it as an image, or perform other actions depending on the environment you are using.
* What we usually do is write plt.show() at the end of every visualisation that we craete in code, so it will directly take us to the visualisation without giving us the normal output. 
* It will basically hide the output so that we can directly see the visualisation.

### <code style="background:yellow;color:black">pie() function:</code>

* **In Matplotlib, the <code style="background:yellow;color:black">pie() function</code> is used to create a pie chart.**

* **The pie() function takes the following arguments:**

    * **x:** A list or array of numeric values representing the sizes of the pie slices. These values should be non-negative and should sum to 100% or 1 if you want the pie to represent percentages.

    * **labels (optional):** A list of labels for each pie slice. If provided, it should have the same length as x.

    * **colors (optional):** A list of colors for the pie slices. If not provided, Matplotlib will use its default color scheme.

    * **autopct (optional):** A string or function to specify how the autopct text should be displayed on each slice. It can display the percentage, the actual value, or a custom format.

    * **startangle (optional):** The angle at which the first slice should start (default is 0 degrees, which is typically at 12 o'clock).

    * **counterclock (optional):** If set to True (the default), the slices will be drawn counterclockwise. If set to False, they will be drawn clockwise.

    * **shadow (optional):** If set to True, the pie chart will have a shadow effect.

    * **explode (optional):** A list or array of values specifying the fraction of the radius with which to offset each slice from the center. This is used to "explode" one or more slices from the center of the pie chart or how "out" that slice will go from the pie chart. It takes n values for n slices.

```python
data.head()
```

```python
sales_data = data[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]]

sales_data
```

```python
sales_data.T
```

```python
sales_data.T.sum(axis = "columns")
```

```python
region_sales = sales_data.sum(axis = "rows")

region_sales
```

```python
plt.pie(
    region_sales,
    labels = region_sales.index,
    startangle = 90,
)

plt.show()
```

```python
plt.pie(
    region_sales,
    labels = region_sales.index,
    startangle = 90,
    explode = (0.2, 0, 0, 0),
    shadow = True,
    autopct = '%1.1f%%'
)

plt.show()
```

Lets suppose we want to do univariate analysis on year column, like how many video games were published in a year. We can plot a <code style="background:yellow;color:black">HISTOGRAM</code> for this. 

**In Matplotlib, the <code style="background:yellow;color:black">hist() function</code> is used to create a histogram, which is a graphical representation of the distribution of a dataset. A histogram divides the data into discrete "bins" or intervals and displays the frequency or count of data points that fall into each bin. This provides a visual representation of the data's underlying distribution.**

**Here are some of the most commonly used arguments:**

* **x:** The dataset or array of values for which you want to create a histogram.

* **bins (optional):** This argument specifies the number of bins or the bin edges. It can be an integer representing the number of bins or a list of bin edges.

* **range (optional):** You can specify the range of data values to be considered. This argument allows you to select a subset of the data for the histogram.

* **density (optional):** If set to True, the histogram will be normalized such that the integral of the histogram is equal to 1, effectively showing a probability density.

* **color (optional):** You can specify the color of the bars in the histogram.

* **label (optional):** A label for the histogram, which can be used in a legend when plotting multiple histograms.

* **histtype (optional):** This argument determines the type of histogram to plot. Common values include 'bar' (regular bar chart), 'step' (unfilled step plot), 'stepfilled' (filled step plot), and more.

* **orientation (optional):** If set to 'horizontal', the histogram will be plotted horizontally.

* **log (optional):** If set to True, the y-axis will be displayed on a logarithmic scale.

* **stacked (optional):** If True, multiple histograms will be stacked on top of each other when plotting multiple datasets.

Now to find out how many video games were published each year, aggregation is required. But hist() function does it automatically, we just have to provide year column. It will automatically create bins and create a histogram. BY default matplotlib created 10 bins for our entire dataset, however we can change those bins.

From this histogram, we can see that video games became very popular in year 2010 or first decade of 2000. Before that and after that it wasnt and is not popular that much.

```python
plt.hist(data["Year"])
plt.show()
```

KDE Plots stands for Kernel Density Estimate and are very similar to histograms, its just that KDE Plots are much smoother versions of histograms. Basically we take the center points of each histogram bar, join all those center points and a curve is generated. This curve is our KDE plot. REFER NOTES FOR MORE CLARITY. We can create KDE plots in various programming libraries, including Seaborn and Matplotlib.  

**In Seaborn, we can use the <code style="background:yellow;color:black">kdeplot() function</code> to create a KDE plot, and it allows us to customize the appearance and bandwidth of the plot.**

**Here are some of the most commonly used arguments:**

* **data:** The data to create the KDE plot for, typically a one-dimensional array-like object.

* **data2 (optional):** If you want to visualize the relationship between two variables, you can provide a second dataset, and kdeplot() will generate a bivariate KDE plot.

* **fill (optional):** If set to True, it fills the area under the KDE curve, creating a shaded region.

* **vertical (optional):** If set to True, the plot is oriented vertically.

* **kernel (optional):** Specifies the kernel to be used for estimation. Common options include "gau" (Gaussian), "cos" (cosine), "biw" (biweight), and others.

* **legend (optional):** If set to True, a legend is added to the plot.

* **cumulative (optional):** If set to True, it plots a cumulative distribution function (CDF) instead of the probability density.

* **common_norm (optional):** If set to True, the densities are normalized to have the same area under the curve.

* **color (optional):** Specifies the color of the KDE plot.

* **ax (optional):** Allows you to specify the axes on which to create the KDE plot if you have multiple subplots.

For more on KDE plots refer the following links:
