## <code style="background:yellow;color:black">1. Box Plot:</code>

* **<code style="background:yellow;color:black">A box plot, also known as a box-and-whisker plot,</code> is a statistical visualization tool that displays the summary of a set of data values. It provides a graphical representation of the distribution of the data based on its five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.**
<br>

**Here's a breakdown of the key components of a box plot:**

* **Box:** The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The box is divided into two parts by the median (Q2).

* **Whiskers:** The whiskers extend from the box to the minimum and maximum values within a certain range. The range can be calculated using various methods, and outliers may be plotted individually.

* **Outliers:** Individual data points that fall outside the range covered by the whiskers are considered outliers and are sometimes plotted separately.
***
### <code style="background:yellow;color:black">Where will we use box plot?</code> 

**Box plots are useful for identifying the central tendency, spread, and skewness of the data distribution.** They are particularly valuable for comparing the distributions of different groups or datasets. Box plots are commonly used in exploratory data analysis and are supported by various data visualization libraries in Python, including Matplotlib and Seaborn.

### <code style="background:yellow;color:black">boxplot() function:</code>

**The <code style="background:yellow;color:black">sns.boxplot() function</code> is specifically designed for creating box plots. It takes as input one or more variables and produces a box plot for each variable, allowing for visual comparison of their distributions.** 

The sns.boxplot function in Seaborn has several parameters that allow us to customize the appearance and behavior of the box plot. **<code style="background:yellow;color:black">re are some of the key parameters:</code>**

* **x, y:** Variables that specify the data for the x and y axes. We can use either x or y or both, depending on how we want to arrange your data.

* **hue:** Variable that produces separate boxes for different levels of the categorical variable specified.

* **data:** DataFrame or long-form data. The data to plot; it can be a DataFrame or any other long-form input.

* **order, hue_order:** Lists of strings that control the order of levels for categorical variables. If not specified, the levels are inferred from the data.

* **orient:** Orientation of the plot. If "v" (default), the boxes are vertical. If "h", the boxes are horizontal.

* **color, palette:** Color for all elements or a palette to use for coloring different levels of the hue variable.

* **width:** Width of the boxes. We can specify a scalar value or a Pandas object with the same length as the input data.

* **linewidth:** Width of the lines outlining the boxes.

* **notch:** Indicate whether to draw a notched box plot. The notches represent the confidence interval around the median.
***

## <code style="background:yellow;color:black">2. Bivariate Analysis</code>

* Bivariate analysis involves the analysis of two variables to determine the empirical relationship between them. 
* It aims to understand the association, correlation, or interaction between the two variables. 
* Bivariate analysis can be useful in exploring and confirming patterns and relationships in data.
* Basically, iit is all about how does one variable change with change in another variable.

**Common methods of bivariate analysis include:**

* **Scatter Plot:** A graphical representation of the relationship between two continuous variables. Each point represents an observation.

* **Correlation Coefficient:** Measures the strength and direction of the linear relationship between two variables. Pearson's correlation coefficient (r) is commonly used for this purpose.

* **Regression Analysis:** Examines the relationship between a dependent variable and one independent variable. Simple linear regression is a common form, which fits a line to the data points.

* **Cross-tabulation (Contingency Table):** A table that displays the frequency distribution of variables. It is often used for categorical data to examine the relationship between two categorical variables.

* When we combine 2 line plots in one chart, label becomes very important. 
* Like the code below, when we provide label parameters, seaborn will create a legend. 
* We dont need to write a seperate command for legend. 

```python
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = baseball, 
             label = "Baseball")
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = ice_hockey, 
             label = "Ice Hockey")
```

* There is another way in matplotlib to show legend, but order in which the plots are plotted needs to match. 
* So remembering order is not preferred because there might be several line plots in single chart, so instead we can pass label arguments in the sns.lineplot() function itself. 

```python
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = baseball)
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = ice_hockey)
plt.legend(["Baseball", "Ice Hockey"]) # Order needs to match
```

* We can use the plt.legend() function to change the positioning of the legend on the chart. 
* The combinations are vague so we have to refer to the documentation to confirm this. 

```python
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = baseball, 
             label = "Baseball")
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = ice_hockey, 
             label = "Ice Hockey")
plt.legend(loc = "lower right")

# upper center
# lower center
# upper right
# lower left
# .
# .
# .
```

* Legend location can also be specified as coordinates and that has to be passed as a tuple. 

* Location in the legend can be specified in two ways:
    1. Coordinates
    2. String Location

```python
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = baseball, 
             label = "Baseball")
sns.lineplot(x = "Year", 
             y = "NA_Sales", 
             data = ice_hockey, 
             label = "Ice Hockey")
plt.legend(loc = (-0.5, 0.5))
```

* We can specify 2 more arguments in legend. 
* One is number of columns and second is whether we want frame on or not. 
* Number of columns will specify within the legend how many columns will be. frameon = False removes the border.

```python
sns.lineplot(x = 'Year', 
             y = 'NA_Sales', 
             data = ice_hockey, 
             label = "Ice Hockey")
sns.lineplot(x = 'Year', 
             y = 'NA_Sales', 
             data = baseball, 
             label = "Baseball")
plt.legend(
    loc='lower right',
    ncol = 2, # Number of columns within the legend
    frameon = False 
)
plt.show()
```

## <code style="background:yellow;color:black">3. Scatter Plot</code>

A scatter plot is a type of data visualization that shows the relationship between two continuous variables. Each point on the scatter plot represents an observation in the dataset, with one variable determining the position on the x-axis and the other variable determining the position on the y-axis. Scatter plots are particularly useful for identifying patterns, trends, correlations, and outliers in the data.

**Key Features of a Scatter Plot**

* **Axes:** The horizontal axis (x-axis) represents one variable, and the vertical axis (y-axis) represents the other variable.
* **Data Points:** Each point on the scatter plot represents a pair of values from the two variables.
* **Trends:** Patterns or trends can often be observed, such as a positive or negative correlation.
* **Clusters:** Groupings of points can indicate clusters or subgroups within the data.
* **Outliers:** Points that fall far from the general pattern can be identified as outliers.<br><br><br>

* When it comes to variance or the change of one variable with respect to another, one more plot comes handy and thats called a Scatter Plot. 
* What line plot does is it generalises the trend or it gives a linear line. 
* But whenever we talk about correlations between different variables, correlations can come in varied structures. REFER NOTES HERE FOR CORRELATION CLARITY AND DIAGRAM. 
* In some cases, we have a strong linear correlation between both the variables, while in some other case, there is not so strong (low to moderate) linear correlation. 
* In some other case, there can be no correlation between 2 variables because these 2 variables are completely unrelated to one another. It could be a strong non-linear correlation as well. 
* So we saw various correlation cases between 2 variables.
